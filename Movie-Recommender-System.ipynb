```python
import numpy as np
import pandas as pd 
import os
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace= True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

# we would take top 3 cast 

```python
import ast

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
import ast
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

 This converts string in individual strings

```python
movies['overview']= movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

This code is for removing space between string

```python
movies['genres']= movies['genres'].apply(lambda x :[i.replace(" ","") for i in x])
movies['keywords']= movies['keywords'].apply(lambda x :[i.replace(" ","") for i in x])
movies['cast']= movies['cast'].apply(lambda x :[i.replace(" ","") for i in x])
movies['crew']= movies['crew'].apply(lambda x :[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']= movies['overview']+movies['genres']+movies['keywords']+movies['crew']+movies['cast']
```

```python
movies.head()
```

```python
new_df= movies[['movie_id','title','tags']]
```

```python
new_df
```

Now we would convert Tags List in string

```python
new_df['tags'][0]
```

```python
new_df.loc[:, 'tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

change in lower case letter

```python
new_df.loc[:, 'tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df.loc[:, 'tags']= new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector= cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vector) 
# distance from another movie
```

```python
def recommend(movie):
    movie_index= new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
